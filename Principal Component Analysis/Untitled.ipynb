```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_X)
```

```python
pc_df = pca.transform(scaled_X)
```

```python
sns.scatterplot(x = pc_df[:, 0], y = pc_df[:, 1]);
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer.target
```

```python
sns.scatterplot(x = pc_df[:, 0], y = pc_df[:, 1], c = cancer.target);
```

```python
pca.components_
```

```python
df_components = pd.DataFrame(data = pca.components_, index = ['pc1', 'pc2'], columns = df.columns )
```

```python
df_components
```

```python
plt.figure(figsize = (16, 6))
sns.heatmap(df_components, annot=True);
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit_transform(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1, 30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)

    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1, 30), explained_variance);
```
