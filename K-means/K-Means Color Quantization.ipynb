# K-Means Colour Quantization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as ig
```

```python
image_array = ig.imread('../DATA/palm_trees.jpg')
```

```python
image_array.shape 
# (H, W, C)
```

```python
plt.figure(dpi=200)
plt.imshow(image_array);
```

```python
# (H, W, C)---> 2D (H*W, C)
```

```python
(h, w, c) = image_array.shape
```

```python
image_2d_array = image_array.reshape(h*w, c)
```

```python
image_2d_array
```

```python
image_2d_array.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 6)
```

```python
labels = model.fit_predict(image_2d_array)
```

```python
len(labels)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
image_2d_array
```

```python
quantized_image = np.reshape(rgb_codes[labels], (h, w, c))
```

```python
plt.figure(dpi=200)
plt.imshow(quantized_image);
```
