```python
def display_catogories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(data = data, x='X1', y = 'X2', hue=labels, palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters = 3)
```

```python
display_catogories(kmean, blob)
```

```python
kmean = KMeans(n_clusters=2)
```

```python
display_catogories(kmean, moon)
```

```python
display_catogories(kmean, circle)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN()
```

```python
display_catogories(db, blob)
```

```python
db1 = DBSCAN(eps= 0.15)
```

```python
display_catogories(db1, moon)
```

```python
display_catogories(db1, circle)
```
